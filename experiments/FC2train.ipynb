```python

# Convert the relevant columns b,z,eta,i,x,K to a PyTorch tensor P_s for parent, with shape (path,N,6).
# path is number of paths, N is number of unique IDs in each path (choose the biggest one).
path_num = df2['path'].nunique()
N = df2.groupby('path')['ID'].nunique().max()
branch_num = 2  # number of children branches per parent
P_s = torch.zeros((path_num, N, 5), dtype=torch.float32)
K_parent = torch.ones((path_num, N, 1), dtype=torch.float32)

alive_mask = torch.zeros((path_num, N), dtype=torch.bool)
entry_mask = torch.zeros((path_num, N, branch_num), dtype=torch.bool)
for i, (path, group) in enumerate(df2.groupby('path')):
    vals = torch.tensor(group[['b', 'z', 'ETA', 'i', 'x']].to_numpy(dtype=np.float32))
    P_s[i, : len(group)] = vals
    K_parent[i, : len(group), 0] = torch.tensor(group['K'].to_numpy(dtype=np.float32))
    alive_mask[i, : len(group)] = torch.tensor(group['b'].notna().to_numpy(dtype=np.float32))
    entry_mask[i, : len(group), 0] = torch.tensor(group['Entry_child1'].to_numpy(dtype=np.float32))
    entry_mask[i, : len(group), 1] = torch.tensor(group['Entry_child2'].to_numpy(dtype=np.float32))



```

```python

# Convert the relevant columns b,z,\eta,i,x to a PyTorch tensor Children_s for child1 and child2, with shape (path, N, branch_num, 5) and K tensor K_chilren for child1 and child2, with shape (path, N, 2branch_num, 1)
Children_s = torch.zeros((path_num, N, branch_num, 5), dtype=torch.float32)
K_children = torch.zeros((path_num, N, branch_num, 1), dtype=torch.float32)
for i, (path, group) in enumerate(df2.groupby('path')):
    vals_child1 = torch.tensor(group[['b_child1', 'z_child1', 'ETA_child1', 'i_child1', 'x_child1']].to_numpy(dtype=np.float32))
    vals_child2 = torch.tensor(group[['b_child2', 'z_child2', 'ETA_child2', 'i_child2', 'x_child2']].to_numpy(dtype=np.float32))
    Children_s[i, : len(group), 0, :] = vals_child1
    Children_s[i, : len(group), 1, :] = vals_child2
    K_children[i, : len(group), 0, 0] = torch.tensor(group['K_child1'].to_numpy(dtype=np.float32))
    K_children[i, : len(group), 1, 0] = torch.tensor(group['K_child2'].to_numpy(dtype=np.float32))
```

```python
Children_s[:,:,:,0] = 0 # set b to 0 for children 
```

```python

# make the P_s and Children_s [path,full_N,5] and [path, full_N, 2, 5] tensors, N is 2000 samples, the rows beyond df set to 0 and same to K_parent and K_children. if the original tensor is equal to full ones ,make the full tensor directly

if P_s.shape[1] == full_N:
    P_s_full = P_s
    K_parent_full = K_parent
    Children_s_full = Children_s
    K_children_full = K_children
else:
    P_s_full = torch.zeros((path_num, full_N, 5), dtype=torch.float32)
    P_s_full[:, :P_s.shape[1], :] = P_s
    K_parent_full = torch.zeros((path_num, full_N, 1), dtype=torch.float32)
    K_parent_full[:, :K_parent.shape[1], :] = K_parent

    Children_s_full = torch.zeros((path_num, full_N, 2, 5), dtype=torch.float32)
    Children_s_full[:, :Children_s.shape[1], :, :] = Children_s
    K_children_full = torch.zeros((path_num, full_N, 2, 1), dtype=torch.float32)
    K_children_full[:, :K_children.shape[1], :, :] = K_children
```

```python

# define a function named compute_quantile_features to compute quantile features for b and z
def compute_quantile_features(b_vals, z_vals, quantiles):
    b_quantiles = torch.quantile(b_vals, quantiles)
    z_quantiles = torch.quantile(z_vals, quantiles)
    quantile_features = torch.cat([b_quantiles, z_quantiles], dim=0)
    return quantile_features
```

```python
# make the input tensors for FC2 model which is FC2_input_parent, it is a (path_num, 201), the first 200 are 100 quantile of b and z distribution which are from each path get the parent features b and z and get their quantiles, the last one is x feature
FC2_input_parent = torch.zeros((path_num, 201), dtype=torch.float32)
for i in range(path_num):
    b_vals = P_s_full[i, alive_mask[i], 0]
    z_vals = P_s_full[i, alive_mask[i], 1]
    quantile_features = compute_quantile_features(b_vals, z_vals, quantiles=torch.linspace(0, 1, steps=100))
    x_feature = P_s_full[i, alive_mask[i], 4].mean().unsqueeze(0)  # Use mean x feature
    FC2_input_parent[i, :200] = quantile_features
    FC2_input_parent[i, 200] = x_feature
```

```python
# FIXME: FC2_output broadcast misses branch dimension; spec children need per-branch macro proxy
# define  a torch model first and then put the FC2_input_parent into FC2model to get the output
class FC2Model(torch.nn.Module):
    def __init__(self):
        super(FC2Model, self).__init__()
        self.fc1 = torch.nn.Linear(201, 128)
        self.fc2 = torch.nn.Linear(128, 64)
        self.fc3 = torch.nn.Linear(64, 2)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = torch.relu(self.fc2(x))
        x = self.fc3(x)
        return x

fc2_model = FC2Model()
fc2_output_parent = fc2_model(FC2_input_parent)  # Shape: (path_num, 1)
```

```python
# FIXME: PV_input dims/fields differ from spec (should include fc2 c/K proxy and respect branch dims)
# broadcast fc2_output_parent to shape (path_num, full_N, 2) and multiply with alive_mask to get the final output tensor FC2_output_children
FC2_output_parent = fc2_output_parent.unsqueeze(1).expand(-1, full_N, -1) * alive_mask.unsqueeze(-1)  # Shape: (path_num, full_N, 2)
```

```python
# concat (P_s_full,FC2_output_parent) to get the final input tensor PV_input with shape (path_num, full_N, 7)
PV_input = torch.cat((P_s_full, FC2_output_parent), dim=-1)
```

```python
# define a simple feedforward neural network model,with input size 7 and output size 3, make sure the value of the first 2 dims of output is whether 1 or 0, and the last dim is a float number, make sure the model will not break the gradient flow
class PVModel(torch.nn.Module):
    def __init__(self):
        super(PVModel, self).__init__()
        self.fc1 = torch.nn.Linear(7, 128)
        self.fc2 = torch.nn.Linear(128, 64)
        self.fc3 = torch.nn.Linear(64, 3)

    def forward(self, x):
        h = torch.relu(self.fc1(x))
        h = torch.relu(self.fc2(h))
        logits = self.fc3(h)
        probs = torch.sigmoid(logits[..., :2])          # probabilities for first two dims
        hard_bits = (probs > 0.5).float()                # hard 0/1 values
        hard_bits = hard_bits + (probs - probs.detach()) # straight-through: keeps gradient from probs
        value = logits[..., 2:3]                        # last dim stays continuous
        return torch.cat([hard_bits, value], dim=-1)
```

```python
# input the PV_input except the -3 element into PVModel to get the final output PV_output with shape (path_num, full_N, 3)
pv_model = PVModel()
PV_output = pv_model(PV_input)  # Shape: (path_num, full_N, 3)
```

```python
# update the alive_mask based on PV_output
# expand alive_prob to match entry_mask shape (path_num, full_N, 2)
alive_prob = PV_output[..., 0].clamp(0, 1)
base_alive = alive_mask.float()           # (path_num, full_N, 1)
alive_prefix = torch.cumprod(alive_prob, dim=1)       # stays 1 until first zero per path/step/branch
updated_alive = base_alive * alive_prob               # broadcast to (path_num, full_N, 2)
# if element in updated_alive is nan, fill it with 0
updated_alive = torch.nan_to_num(updated_alive, nan=0.0)
updated_alive = updated_alive.unsqueeze(-1).expand(-1, -1, 2)
entry_flag = entry_mask.float()                         # (path_num, full_N, 2)
alive_mask = updated_alive + entry_flag
```

```python
# compute real lnk and hatc, for parent. for each path, choose all alive parents and compute their lnk and hatc based on PV_output and PV_input. Y = torch.exp(x + z) * K Phi = (1 - self.phi) * (1 + torch.exp(x + z)) * K * bar_z I = bar_i * K * i - bar_z * K + self.delta * K C = Y - I - Phi and then lnK is sum of all alive parents'K and get log, well hatc is sum of all alive parents'C and divide sum of K and get log. which bar_z is the first dimision of PV_output, bar_i is the second dimension of PV_output, and self.phi and self.delta are some constants, set self.phi = 0.1 and self.delta = 0.05 for now. using tensor operations to compute them without for loop.

phi = 0.1
delta = 0.05
bar_z = PV_output[..., 0:1]  # Shape: (path_num, full_N, 1)
bar_i = PV_output[..., 1:2]  # Shape: (path_num, full_N, 1)
Y = torch.exp(PV_input[..., 4:5] + PV_input[..., 1:2]) * K_parent_full  # Shape: (path_num, full_N, 1)
Phi = (1 - phi) * (1 + torch.exp(PV_input[..., 1:2] + PV_input[..., 3:4])) * K_parent_full * bar_z  # Shape: (path_num, full_N, 1)
I = bar_i * K_parent_full * PV_input[..., 3:4] - bar_z * K_parent_full + delta * K_parent_full  # Shape: (path_num, full_N, 1)
C = torch.abs(Y - I - Phi)  # Shape: (path_num, full_N, 1)
masked_K = torch.where(
    updated_alive[...,0:1] > 0,
    K_parent_full,
    torch.zeros_like(K_parent_full)
)

lnk_parent = torch.log(masked_K.sum(dim=1) + 1e-8)

masked_C = torch.where(
    updated_alive[...,0:1] > 0,
    C,
    torch.zeros_like(C)
)

hatc_parent = torch.log(torch.sum(masked_C, dim=1) / (torch.sum(masked_K, dim=1) + 1e-8))  # Shape: (path_num, 1)

```

```python
# compute loss of FC2 model as the mean squared error between lnk_parent and fc2_output_parent[..., 0] and hatc_parent and fc2_output_parent[..., 1]
loss_parent = torch.mean((lnk_parent - fc2_output_parent[..., 0:1])**2) + torch.mean((hatc_parent - fc2_output_parent[..., 1:])**2)
```

```python
# update children_s_full based on alive_mask and PV_output, children_s_full[:,:,:,0] is set to PV_output[...,2] * P_s_full[:,:,2] +  P_s_full[:,:,0] * (1- P_s_full[:,:,2])

masked_b = torch.where(
    updated_alive[...,0] > 0,
    P_s_full[:,:,0],
    torch.zeros_like(P_s_full[:,:,0])
)
masked_bp = torch.where(
    updated_alive[...,0] > 0,
    PV_output[...,2],
    torch.zeros_like(PV_output[...,2])
)

Children_s_full[:,:,0,0] = (masked_bp * Children_s_full[:,:,0,2]  +  masked_b * (1- Children_s_full[:,:,0,2]))
Children_s_full[:,:,1,0] = (masked_bp * Children_s_full[:,:,1,2] +  masked_b * (1- Children_s_full[:,:,1,2] ))
```

```python
# make the input tensors for FC2 model which is FC2_input_children it is a (path_num, 2, 201), the first 200 are 100 quantile of b and z distribution which are from each path get the parent features b and z and get their quantiles, the last one is x feature
FC2_input_children = torch.zeros((path_num, 2, 201), dtype=torch.float32)
for i in range(path_num):
    for j in range(2):
        b_vals = Children_s_full[i, alive_mask[i,:,j].bool(), j, 0]
        z_vals = Children_s_full[i, alive_mask[i,:,j].bool(), j, 1]
        quantile_features = compute_quantile_features(b_vals, z_vals, quantiles=torch.linspace(0, 1, steps=100))
        x_feature = Children_s_full[i, alive_mask[i,:,j].bool(), j, 4].mean().unsqueeze(0)  # Use mean x feature
        FC2_input_children[i, j, :200] = quantile_features
        FC2_input_children[i, j, 200] = x_feature
```

```python
# concat (Children_s_full,FC2_output_children) to get the final input tensor CV_input with shape (path_num, full_N, 2, 7)
CV_input = torch.cat((Children_s_full, FC2_output_children), dim=-1)
```

```python
# input the CV_input except the -3 element into PVModel to get the final output CV_output with shape (path_num, full_N, 3)
CV_output = pv_model(CV_input)
```

```python
# update the alive_mask_children based on CV_output

alive_mask_children = alive_mask.clone()
# update the alive_mask_children based on CV_output
alive_prob_children = CV_output[..., 0].clamp(0, 1)
base_alive_children = alive_mask.float()           # (path_num, full_N, 1)
updated_alive_children = base_alive_children * alive_prob_children               # broadcast to (path_num, full_N, 2)
alive_mask_children = updated_alive_children.unsqueeze(-1)
```

```python
# update K_children_full based on K_children_full and CV_output, for bari which is the second dimision of CV_output, if bari is 1, then K_children_full = K_children_full * (1 + G), else K_children_full = K_children_full ,G is 0.14, keep the gradient flow
G = 0.14
bari = CV_output[..., 1:2]  # Shape: (path_num, full_N, 1)
K_children_full = K_children_full * (1 + G * bari) 
```

```python
# compute real lnk and hatc, for children. for each path, choose all alive children and compute their lnk and hatc based on PV_output and PV_input. Y = torch.exp(x + z) * K Phi = (1 - self.phi) * (1 + torch.exp(x + z)) * K * bar_z I = bar_i * K * i - bar_z * K + self.delta * K C = Y - I - Phi and then lnK is sum of all alive parents'K and get log, well hatc is sum of all alive children's C and divide sum of K and get log. which bar_z is the first dimision of PV_output, bar_i is the second dimension of PV_output, and self.phi and self.delta are some constants, set self.phi = 0.1 and self.delta = 0.05 for now. using tensor operations to compute them without for loop.
phi = 0.1
delta = 0.05
bar_z_children = CV_output[..., 0:1]  # Shape: (path_num, full_N, 2, 1)
bar_i_children = CV_output[..., 1:2]  # Shape: (path_num, full_N, 2, 1)
Y = torch.exp(CV_input[..., 4:5] + CV_input[..., 1:2]) * K_children_full  # Shape: (path_num, full_N, 2, 1)
Phi = (1 - phi) * (1 + torch.exp(CV_input[..., 1:2] + CV_input[..., 3:4])) * K_children_full * bar_z_children  # Shape: (path_num, full_N, 2, 1)
I = bar_i_children * K_children_full * CV_input[..., 3:4] - bar_z_children * K_children_full + delta * K_children_full  # Shape: (path_num, full_N, 2, 1)
C = torch.abs(Y - I - Phi)  # Shape: (path_num, full_N, 2, 1)
lnk_children = torch.log(torch.sum(K_children_full * alive_mask_children[...,0:1], dim=1) + 1e-8)  # Shape: (path_num, 1)
hatc_children = torch.log(torch.sum(C * alive_mask_children[...,0:1], dim=1) / (torch.sum(K_children_full * alive_mask_children[...,0:1], dim=1) + 1e-8))  # Shape: (path_num, 1)
```
