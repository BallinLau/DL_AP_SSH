# Compare: Sample(data_mode='simulate') vs SimulateTS(horizon=2)

This notebook compares outputs from:
- `Sample(..., data_mode='simulate')` (cross-sectional, one-step)
- `SimulateTS(..., horizon=2)` (time-series simulation)

It prints shapes/columns and shows head rows for firm/macro outputs.


```python
import sys
from pathlib import Path
import torch
import pandas as pd


ROOT = Path().resolve()
if not (ROOT / 'config').exists():
    ROOT = ROOT.parent
sys.path.append(str(ROOT))

from config import Config
from data.sample import Sample
from data.simulate_ts import SimulateTS
from experiments.run_utils import build_models

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
Config.DEVICE = device
print('device:', device)
```

```python
# Build models (same as run_multi_episode.py)
models = build_models(device)
```

## Sample(data_mode='simulate')


```python
# Small settings for debug
n_samples = 60
n_paths = 10
group_size = 2
n_branches = 2

sampler = Sample(
    models=models,
    config=Config,
    n_samples=n_samples,
    n_paths=n_paths,
    group_size=group_size,
    branch_num=n_branches,
    data_mode='simulate',
    device=device,
)

df_firm_s, df_macro_s = sampler.build_fc2_df()

print('Sample simulate: df_firm shape:', df_firm_s.shape)
print('Sample simulate: df_firm columns:', df_firm_s.columns.tolist())
display(df_firm_s.head(10))

print('Sample simulate: df_macro shape:', df_macro_s.shape)
print('Sample simulate: df_macro columns:', df_macro_s.columns.tolist())
display(df_macro_s.head(10))
```

## SimulateTS(horizon=2)


```python
sim = SimulateTS(
    models=models,
    config=Config,
    n_paths=n_paths,
    group_size=group_size,
    branch_num=n_branches,
    horizon=1,
    enable_entry=False,
    enable_exit=False,
    device=device,
)

df_firm_ts, df_macro_ts = sim.simulate()

print('SimulateTS: df_firm shape:', df_firm_ts.shape)
print('SimulateTS: df_firm columns:', df_firm_ts.columns.tolist())
display(df_firm_ts.head(10))

print('SimulateTS: df_macro shape:', df_macro_ts.shape)
print('SimulateTS: df_macro columns:', df_macro_ts.columns.tolist())
display(df_macro_ts.head(10))
```

## Quick comparison summary


```python
def summarize(df, name):
    print(f'[{name}] rows={len(df)}, cols={len(df.columns)}')

summarize(df_firm_s, 'sample df_firm')
summarize(df_macro_s, 'sample df_macro')
summarize(df_firm_ts, 'ts df_firm')
summarize(df_macro_ts, 'ts df_macro')

common_firm_cols = sorted(set(df_firm_s.columns).intersection(df_firm_ts.columns))
common_macro_cols = sorted(set(df_macro_s.columns).intersection(df_macro_ts.columns))

print('common firm cols:', common_firm_cols)
print('common macro cols:', common_macro_cols)
```

## Derive x_t / x_t1 / Hatcf_t / LnKF_t from df_macro_ts


```python
# Build sdf-style macro pairs from df_macro_ts (with Hatc/LnK at t and t+1)
parent = df_macro_ts[df_macro_ts['branch'] == -1][
    ['path', 't', 'x', 'hatcf', 'lnkf', 'Hatc', 'LnK']
].rename(columns={
    't': 't_parent',
    'x': 'x_t',
    'hatcf': 'Hatcf_t',
    'lnkf': 'LnKF_t',
    'Hatc': 'Hatc_t',
    'LnK': 'LnK_t',
})

child = df_macro_ts[df_macro_ts['branch'] >= 0][
    ['path', 't', 'branch', 'x', 'Hatc', 'LnK']
].rename(columns={
    'x': 'x_t1',
    'Hatc': 'Hatc_t1',
    'LnK': 'LnK_t1',
})

child['t_parent'] = child['t'] - 1
df_sdf_ts = child.merge(parent, on=['path', 't_parent'], how='left')

df_sdf_ts = df_sdf_ts[[
    'path', 't', 'branch',
    'x_t', 'x_t1',
    'Hatcf_t', 'LnKF_t',
    'Hatc_t', 'LnK_t',
    'Hatc_t1', 'LnK_t1'
]]

print('df_sdf_ts shape:', df_sdf_ts.shape)
display(df_sdf_ts.head(10))
```
